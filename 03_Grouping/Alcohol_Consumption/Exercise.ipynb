### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
drinks
```

### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby("continent").agg({'beer_servings' : 'mean'}).idxmax()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby("continent").agg({'wine_servings' : 'sum', 'total_litres_of_pure_alcohol' : 'sum'})
```

```python
drinks.groupby('continent').wine_servings.describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby(["continent", "country"]).agg({
    'wine_servings' : 'mean', 
    'total_litres_of_pure_alcohol' : 'mean', 
    'beer_servings' : 'mean',
    'spirit_servings' : 'mean'
    })
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby(["continent", "country"]).agg({
    'wine_servings' : 'mean', 
    'total_litres_of_pure_alcohol' : 'median', 
    'beer_servings' : 'mean',
    'spirit_servings' : 'mean'
    })
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
# drinks[drinks["spirit_servings"] == drinks["spirit_servings"].min()]
drinks["spirit_servings"].min()
```

```python
# drinks[drinks["spirit_servings"] == drinks["spirit_servings"].max()]
drinks["spirit_servings"].max()
```

```python
# drinks[drinks["spirit_servings"] == drinks["spirit_servings"].mean()]
drinks["spirit_servings"].mean()
```
